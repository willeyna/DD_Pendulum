# Simple Pendulum Equations of Motion

For a simple pendulum with mass $m$ and length $l$, we can model its motion as

$$ml \frac{d^2\theta}{dt^2} +mg \sin(\theta) = 0$$

When we assume a linear air resistance acts on the pendulum _and_ add an external force acting on the pendulum along its direction of motion, our equation of motion becomes as follows:

$$ ml \frac{d^2\theta}{dt^2} + b\frac{d\theta}{dt} +mg \sin(\theta) = F^{\text{ext}}$$

This ODE is non-linear and has no general analytic solution. We can, however, find the motion of the system by numerically integrating the equation of motion as two coupled differential equations: 


$$\frac{d \omega}{dt} = \frac{1}{ml}F^{\text{ext}} - \frac{b}{ml} \omega - g \sin(\theta)$$

$$ \frac{d \theta}{dt}  = \omega$$

---

## Scaling the Equation

We can opt to rewrite the above equation as a dimensionlesss one through a series of subsitutions and re-writings.

First, let the following definitions hold:

$$\omega_0 = \sqrt{\frac{g}{l}}$$

$$\tau = \omega_0 t$$

$$\tilde{\omega} = \frac{\omega_\text{ext}}{\omega_0} \ \implies \omega_\text{ext} t = \tilde{\omega} \tau$$

We can subsitute these values into our equation of motion to obtain: 

$$ \omega_0^2 l \frac{d^2\theta}{d \tau^2} + \frac{b \omega_0}{m}\frac{d\theta}{d \tau} +g \sin(\theta) = \frac{F^{\text{ext}}}{m}$$

$$ \frac{d^2\theta}{d \tau^2} + \frac{b}{m l \omega_0}\frac{d\theta}{d \tau} + \sin(\theta) = \frac{F^{\text{ext}}}{m \omega_0^2}$$

Denote $\gamma :=  \frac{b}{2m l \omega_0}$ and we obtain $$ \frac{d^2\theta}{d \tau^2} + 2\gamma\frac{d\theta}{d \tau} + \sin(\theta) = \frac{F^{\text{ext}}}{m \omega_0^2}$$


Lastly, if we assume our force $F^{\text{ext}}$ is of the form $F^{\text{ext}} = F_0 \sin (\omega_\text{ext} t)$ and let $\tilde{F_0} = \frac{F_0}{m \omega_0^2}$, then we arrive at the following:

$$\frac{d^2 \theta}{d \tau^2} + 2\gamma\frac{d\theta}{d \tau} + \sin(\theta)= \tilde{F_0}sin(\tilde{\omega}\tau)$$

---

This scaling is done as an exercise, and the numeric solutions to follow come from the original coupled set of differential equations.

```python
#common imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
#custom solver class
from solver import *
%matplotlib inline

import seaborn as sns
sns.set(context='paper', style='darkgrid', rc={'figure.facecolor':'white'}, font_scale=1.2)
```

```python
def Forces(r, v, t, i, m):
    
    #### CONSTANTS 
    l = 1
    m = 1
    g = 1
    b = 1
    F_ext = 0 
    
    F = np.zeros_like(r[0,:])
    
    #set each dimension's net force
    Fx = (1/(m*l))*F_ext - (b/(m*l))*v[i,0] - g*np.sin(r[i,0])
    Fy = 0
    Fz = 0
    
    F_PACK = [Fx, Fy, Fz] 
    
    #pulls the correct dim of forces to return
    for dim in range(F.shape[0]):
        F[dim] = F_PACK[dim]
    
    return F

fig, axes = plt.subplots(2,1, figsize = (15,10))

phi0 = [0.2]
omega0 = [0]

damped_pendulum = Solver(phi0, omega0, Forces)
damped_pendulum.EulerCromer(t0 = 0, tf = 10, dt = .1)
t = damped_pendulum.t
phi = damped_pendulum.position
omega = damped_pendulum.velocity

axes[0].plot(t,phi, label = "N=100")
axes[1].plot(phi,omega, label = "N=100")

damped_pendulum.EulerCromer(t0 = 0, tf = 10, dt = .01)
t = damped_pendulum.t
phi = damped_pendulum.position
omega = damped_pendulum.velocity

axes[0].plot(t,phi, label = "N=1000")
axes[1].plot(phi,omega, label = "N=1000")

damped_pendulum.EulerCromer(t0 = 0, tf = 10, dt = .001)
t = damped_pendulum.t
phi = damped_pendulum.position
omega = damped_pendulum.velocity

axes[0].plot(t,phi, label = "N=10000")
axes[1].plot(phi,omega, label = "N=10000")

axes[0].legend()
fig.suptitle(r'$\nu$ = 1 with Varying Meshpoints')
```

# Error Analysis Code

```python
ri = [1]
vi = [0]
m = 1
gamma = .2
Omegatilde = 2
Ftilde = 1

steps = 1/(10**(np.arange(2,13,1)))
tf = 1

def pend(y, t, gamma, Omegatilde, Ftilde):
    theta, omega = y
    dydt = [omega, (-2*gamma*omega-np.sin(theta)+Ftilde*np.cos(t*Omegatilde))]
    return dydt

y0 = [r[0][0],v[0][0]]

euler_vals = np.zeros_like(steps)
odeint_vals = np.zeros_like(steps)

for i,step in enumerate(steps):
    r,v,a,t = Euler(ri,vi,m,step, tf = tf)
    sol = odeint(pend, y0, t, args=(gamma, Omegatilde, Ftilde))
    r_odeint, v_odeint = sol[:,0], sol[:,1]
    euler_vals[i] = r.flatten()[-1]
    odeint_vals[i] = r_odeint[-1]
```

```python
rel_err = np.log10(np.abs((euler_vals - odeint_vals)/odeint_vals))
logt = np.log10(steps)

#m,b = np.polyfit(logt, rel_err, 1)
plt.plot(logt, rel_err)
print(f'y = {m}x + {b}')
```
